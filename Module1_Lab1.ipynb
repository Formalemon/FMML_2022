### Answer 0:

#### List
1. append(item) - adds an item to the end of the list
2. extend(iterable) - adds all elements of an iterable to the end of the list
3. insert(index, item) - inserts an item at a specific index
4. remove(item) - removes the first occurrence of an item from the list
5. reverse() - reverses the order of the elements in the list

#### String
1. upper() - returns a string in all uppercase letters
2. lower() - returns a string in all lowercase letters
3. split(separator) - splits a string into a list of substrings, using the specified separator as the delimiter
4. join(iterable) - concatenates elements of an iterable with the string as separator
5. replace(old, new) - replaces all occurrences of a substring in a string with another substring

#### Dictionary
1. keys() - returns a list of all keys in the dictionary
2. values() - returns a list of all values in the dictionary
3. items() - returns a list of tuples, where each tuple contains a key-value pair from the dictionary
4. get(key, default) - returns the value of a key if it exists in the dictionary, otherwise returns the default value
5. pop(key) - removes and returns the value associated with the given key from the dictionary

#### Tuple
1. count(item) - returns the number of times an item appears in the tuple
2. index(item) - returns the index of the first occurrence of an item in the tuple
3. +(tuple) - concatenates two tuples
4. *(n) - returns a new tuple with n copies of the original tuple
5. tuple(list) - converts a list into a tuple

#### Set
1. add(item) - adds an item to the set
2. update(iterable) - adds all elements of an iterable to the set
3. remove(item) - removes an item from the set, raises KeyError if item is not in the set
4. discard(item) - removes an item from the set if it is present, does nothing if item is not in the set
5. pop() - removes and returns an arbitrary element from the set, raises KeyError if the set is empty
6. clear() - removes all elements from the set

```python
def check_even_list(num_list):
    
    even_numbers = []
    
    # Go through each number
    for number in num_list:
        # Once we get a "hit" on an even number, we append the even number
        if number % 2 == 0:
            even_numbers.append(number)
        # Don't do anything if its not even
        else:
            pass
    # Notice the indentation! This ensures we run through the entire for loop    
    return even_numbers
```

#### Question 2 : 
##### Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.

##### work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]

##### Answer : the female worker that worked the most hours is Penny and she should be paid 18400

### Answer 1:

```python
def reverse_and_append(A, B):
    A_reversed = A[::-1] # reverse string A
    return B + A_reversed # append A_reversed to B and return the final string

# Usage:
reverse_and_append("hello", "world")
```

### Answer 2:

```python
def max_hours_female_worker(work_hours):
    female_hours = {}
    for name, hours, gender in work_hours:
        if gender == 'F':
            female_hours[name] = hours
    max_hours = max(female_hours.values())
    for name, hours in female_hours.items():
        if hours == max_hours:
            pay = hours * 20
            print(f"The female worker who worked the most hours is {name} and she should be paid ${pay}.")
            break

# Given Data:
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]

# Usage:
max_hours_female_worker(work_hours)
```

```python
a = np.array([1,1,2,3,4,5,5,6,1]) ## np.array converts given list to array

b = a>1 ## important comparison operation, where frequently used in manipulation and image processing.

print(b)
print(a[b]) ## [printing only those values in a which are greater than 1]
```

```python
a_range = np.arange(10,19).reshape(3,3) ## create a 3x3 array with values in range 10-19
a_range
```

```python
## Indexing in arrays works same as that of list

a_range[0] # printing all the columns of first row
```

```python
a_range[:,2] #printing all the rows of second column
```

```python
iden = np.eye(3) #idnetity matrix of given size
iden
```

```python
## adding two matrices
summed = a_range + iden
summed
```

```python
### arrays support normal matrix multiplication that you are used to, point-wise multiplication
### and dot product as well.

mul = a_range@iden ## normal multiplication
mul
```

```python
## point wise multiplication
p_mul = a_range * iden
p_mul
```

```python
## Transpose of a matrix.

mtx_t = mul.T
mtx_t
```

```python
### Here we are changing the values of last row of the transposed matrix.
### basically point wise multiplying the values of last row with 1,2 and 3

mtx_t[2] = mtx_t[2]*[1,2,3] ## indexing, point wise multiplication and mutation of values
mtx_t
```

```python
## Just like the greater than 1 (a>1) example we saw earlier.
## here we are checking if the elements are divisible by 2 (%), and if they are, then replace by 0.

mtx_t[(mtx_t % 2 == 0)] = 0 ## convert even elements of the matrix to zero.
mtx_t
```

```python
a = np.array([[1, 2],
              [3, 4],
              [5, 6]])

b = np.array([[7, 8, 9],
              [10, 11, 12]])

# Compute the dot product of a and b if possible
if a.shape[1] != b.shape[0]:
    raise ValueError("Dot product not possible: number of columns in matrix a must equal number of rows in matrix b")
else:
    c = np.dot(a, b)

# Print the result
print(c)
```
